```python
link = 'https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DA0101EN/automobileEDA.csv'

import pandas as pd
df = pd.read_csv(link)
df.head()
```

**Correlation**

---



```python
df_numeric = df.select_dtypes(include=['number'])

df_numeric.corr()  #to see the correlation among the numeric features
```

```python
df[['engine-size', 'city-mpg', 'bore', 'stroke', 'horsepower', 'price']].corr()

#to select some of correlation
```

**Statistic Method**

---

To find the correlation and the P-value

```python
import scipy.stats as st #static method

correlation, p_value = st.pearsonr(df['engine-size'], df['price'])

print('The correlation is:', correlation, 'and the p_value is:', p_value)  #using pearson correlation to see the correlation
```

```python
#9.25 divide by 164zeroes
```

```python
correlation, p_value = st.pearsonr(df['city-mpg'], df['price'])

print('The correlation is:', correlation, 'and the p_value is:', p_value)
```

```python
#check missing value

df['stroke'].isnull().sum()
```

```python
#there still 4 missing value in column stroke
```

```python
import numpy as np

df['stroke'].replace(np.nan, df['stroke'].mean(), inplace=True)

```

```python
correlation, p_value = st.pearsonr(df['stroke'], df['price'])

print('The correlation is:', correlation, 'and the p_value is:', p_value)

```

**GroupBy**

---



```python
df1 = df[['drive-wheels', 'body-style', 'price']]

df1_group = df1.groupby(['drive-wheels', 'body-style'], as_index=False).mean()

df1_group
```

```python
#df1 = df[['drive-wheels', 'body-style', 'price']]

#df1_group = df1.groupby(['drive-wheels', 'body-style'], as_index=True).mean()

#df1_group #it more harder to read and understandable
```

**Pivot Table**

---



```python
#we can convert group by to a pivot table

df1_pivot = df1_group.pivot(index='body-style', columns='drive-wheels')
#row is called index
df1_pivot
```

**Heat Map**

---



```python
sns.heatmap(df1_pivot)
```

```python
#you can have value on each column

sns.heatmap(df1_pivot, annot=True) #default anoot is none
```

```python
#you can have value on each column

sns.heatmap(df1_pivot, annot=True, cmap='RdBu') #default anoot is none
```

```python
#using heat map to find correlation

sns.heatmap(df_numeric.corr(), annot=True, cmap='RdBu')
```
