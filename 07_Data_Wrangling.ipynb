**Simple Features Scalling**

---



```python
#Simple features scalling

mydata['width'].head()
```

```python
mydata[['width', 'length', 'height']].head()
```

```python
#normalize it

mydata['width'] = mydata['width']/mydata['width'].max()

mydata[['width', 'length', 'height']].head()
```

```python
mydata['width'].describe()
```

**Min - Max**

---


Can use this formula or use min max scaler

```python
#min - max

mydata['length'] = (mydata ['length'] - (mydata['length'].min()))/(mydata['length'].max() - mydata['length'].min())

mydata[['width', 'length', 'height']].head()
```

```python
mydata['length'].describe()
```

**Z-Score**

---



```python
#z-score

mydata['height'] = (mydata['height'] - mydata['height'].mean()) / mydata['height'].std()

mydata[['width', 'length', 'height']].head()
```

```python
mydata[['width', 'length', 'height']].describe()
```

**Data Binning**

---

* It is better to separate the og data and the data that you do for cleaning since data binning and few categorical features are best read with og data

```python
#to have a range between small group for better understanding
```

```python
mydata['price'].describe(include= 'all')
```

```python
price_bins = np.linspace(mydata_new1['price'].min(), mydata_new1['price'].max(), num=4)

print(price_bins)
```

```python
#need to group the price

price_bins = np.linspace(mydata_new1['price'].min(), mydata_new1['price'].max(), num=4)

print (price_bins)


price_groups = ['low Price', 'Mid Price', 'High Price']

mydata_new1['price_bins'] = pd.cut(mydata_new1['price'], bins=price_bins, labels=price_groups, include_lowest=True)

mydata_new1[['price', 'price_bins']].head(20)

#include lowest to avoid null value

```

```python
mydata_new1['price_bins'].value_counts()
```
