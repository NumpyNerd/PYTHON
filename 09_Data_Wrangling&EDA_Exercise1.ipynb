```python
#normalize it using simple features scalling

#normalize it

mydata['Alcohol-drinks-day'] = mydata['Alcohol-drinks-day']/mydata['Alcohol-drinks-day'].max()

mydata[['Alcohol-drinks-day']].head(10)
```

```python
mydata['Waist-circum'] = mydata['Waist-circum']/mydata['Waist-circum'].max()
mydata['Arm-circum'] = mydata['Arm-circum']/mydata['Arm-circum'].max()
mydata['Upper-arm-length'] = mydata['Upper-arm-length']/mydata['Upper-arm-length'].max()
mydata['Upper-leg-length'] = mydata['Upper-leg-length']/mydata['Upper-leg-length'].max()
mydata['Height'] = mydata['Height']/mydata['Height'].max()
mydata['BMI'] = mydata['BMI']/mydata['BMI'].max()
mydata['Weight'] = mydata['Weight']/mydata['Weight'].max()
mydata['Diastollic2-BloodPress'] = mydata['Diastollic2-BloodPress']/mydata['Diastollic2-BloodPress'].max()
mydata['Systolic2-BloodPress'] = mydata['Systolic2-BloodPress']/mydata['Systolic2-BloodPress'].max()
mydata['Diastollic1-BloodPress'] = mydata['Diastollic1-BloodPress']/mydata['Diastollic1-BloodPress'].max()
mydata['Systolic1-BloodPress'] = mydata['Systolic1-BloodPress']/mydata['Systolic1-BloodPress'].max()
mydata['2yrs-sample-weight'] = mydata['2yrs-sample-weight']/mydata['2yrs-sample-weight'].max()
mydata['Marital-status'] = mydata['Marital-status']/mydata['Marital-status'].max()

```

Data Binning

```python
#Creating Bins for 'Alcohol-drinks-day'
Alcohol_bins = mydata['Alcohol-drinks-day']

Alcohol_bins = np.linspace(mydata['Alcohol-drinks-day'].min(), mydata['Alcohol-drinks-day'].max(), num=3)

#Printing the bins
print(Alcohol_bins)
```

```python
#Label
Alcohol_groups = mydata['Alcohol-drinks-day']
Alcohol_groups = ['Light Drinker','Heavy Drinker']

#Assigning Data
mydata['Alcohol_bins'] = pd.cut(mydata['Alcohol-drinks-day'],Alcohol_bins, labels = Alcohol_groups, include_lowest=True)

#Binned Data
mydata[['Alcohol-drinks-day','Alcohol_bins']].head(10)
```
